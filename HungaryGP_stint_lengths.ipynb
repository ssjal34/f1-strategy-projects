```python
#Bar Chart version
import fastf1
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Create the cache directory if it doesn't exist
if not os.path.exists('cache'):
    os.makedirs('cache')

fastf1.Cache.enable_cache('cache')

#Load session of the Hungarian Grand Prix Race
year = 2025

gp_name = 'Hungarian Grand Prix'

session = fastf1.get_session(year, gp_name, 'R')

session.load()

#Load in race laps
laps = session.laps

#Group stints by driver and compound
stints = laps[['Driver', 'Stint', 'Compound', 'LapNumber']].groupby(
    ['Driver', 'Stint', 'Compound']
)['LapNumber'].count().reset_index()

stints.rename(columns={'LapNumber': 'StintLength'}, inplace=True)

#Plotting
plt.figure(figsize=(10,6))
sns.barplot(data=stints, x='Driver', y='StintLength', hue='Compound', dodge=True)
plt.title(f'{year} {gp_name} â€“ Stint Lengths by Tyre Compound', fontsize=14, fontweight='bold')
plt.ylabel('Laps in Stint')
plt.xlabel('Driver')
plt.legend(title='Compound')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Gantt Chart version
import fastf1
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Create the cache directory if it doesn't exist
if not os.path.exists('cache'):
    os.makedirs('cache')

fastf1.Cache.enable_cache('cache')

#Load in session of Hungarian Grand Prix Race
year = 2025

gp_name = 'Hungarian Grand Prix'

session = fastf1.get_session(year, gp_name, 'R')

session.load()

laps = session.laps

#Grouping by Drivers and compounds, defining start of a stint and end of a stint
stints = laps.groupby(['Driver', 'Stint','Compound']).agg(
    start_lap = ('LapNumber', 'min'),
    end_lap = ('LapNumber','max'),
    stint_length = ('LapNumber', 'count')
).reset_index()

drivers = stints['Driver'].unique()
drivers.sort()

colors = {'SOFT': 'red', 'MEDIUM': 'yellow', 'HARD': 'white'}

#Plotting
plt.figure(figsize=(16,10), facecolor='black')

for i, driver in enumerate(drivers):
  driver_stints = stints[stints['Driver'] == driver]
  for _, stint in driver_stints.iterrows():
    plt.barh(
        y = driver,
        width = stint['stint_length'],
        left = stint['start_lap'],
        color = colors[stint['Compound']],
        edgecolor = 'black'
    )

plt.xlabel('Lap Number', color='white', fontsize=14)
plt.ylabel('Driver', color='white', fontsize=14)
plt.title('Stint Timeline per Driver - 2025 Hungary GP', fontsize=14, fontweight='bold', color='white')
plt.grid(True, color='white', linestyle='--', linewidth=0.5)
plt.gca().set_facecolor('black')
plt.tick_params(colours='white')

legend_patches = [plt.line2D([0],[0], color=color, lw = 4, label = compound) for compound, color in colors.items()]
plt.legend(handles=legend_patches, facecolor = 'black', edgecolor='white', labelcolor='white', loc='upper right', fontsize=12)
plt.tight_layout()
plt.show()
```
