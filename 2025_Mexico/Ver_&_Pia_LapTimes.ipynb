```python
import fastf1
import pandas as pd
import os

# Enable Cache

if not os.path.exists('cache'):
  os.makedirs('cache')
fastf1.Cache.enable_cache('cache')

# Load session
session = fastf1.get_session(2025, 'Mexico City', 'R')
session.load()

# Pick Verstappen and Piastri
drivers = ['VER', 'PIA']

# Get their laps
laps = session.laps.pick_drivers(drivers)

# Select and format lap number + lap time
for driver in drivers:
    driver_laps = laps.pick_driver(driver)[['LapNumber', 'LapTime']].dropna()
    print(f"\n=== {driver} Lap Times ===")
    for _, row in driver_laps.iterrows():
        print(f"Lap {int(row['LapNumber'])}: {row['LapTime'].total_seconds():.3f} s")
```
